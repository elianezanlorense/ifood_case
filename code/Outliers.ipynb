```python
import pandas as pd
import numpy as np
from statsmodels.stats.proportion import proportions_ztest
import warnings
warnings.filterwarnings('ignore')

print("=== ITEM 1: AN√ÅLISE DA M√âTRICA BIN√ÅRIA - TAXA DE RETEN√á√ÉO ===\n")

# =============================================================================
# AN√ÅLISE DA TAXA DE RETEN√á√ÉO (M√âTRICA BIN√ÅRIA)
# =============================================================================

def analisar_retencao(df):
    """Analisa a taxa de reten√ß√£o com teste de signific√¢ncia"""
    
    # Calcular taxas de reten√ß√£o - FORMA CORRETA
    target_retencao = df[df['is_target']=='target']['retido_jan'].mean()
    control_retencao = df[df['is_target']=='control']['retido_jan'].mean()
    
    lift_absoluto = target_retencao - control_retencao
    lift_relativo = (lift_absoluto / control_retencao) * 100 if control_retencao > 0 else 0
    
    print("--- TAXA DE RETEN√á√ÉO ---")
    print(f"Target: {target_retencao:.4f} ({target_retencao*100:.2f}%)")
    print(f"Control: {control_retencao:.4f} ({control_retencao*100:.2f}%)")
    print(f"Lift absoluto: {lift_absoluto:.4f} ({lift_absoluto*100:.2f}%)")
    print(f"Lift relativo: {lift_relativo:.2f}%")
    
    # Teste de propor√ß√µes Z-test
    target_success = df[df['is_target']=='target']['retido_jan'].sum()
    target_total = df[df['is_target']=='target'].shape[0]
    control_success = df[df['is_target']=='control']['retido_jan'].sum()
    control_total = df[df['is_target']=='control'].shape[0]
    
    print(f"\n--- TESTE DE SIGNIFIC√ÇNCIA ---")
    print(f"Target: {target_success} retidos de {target_total} usu√°rios")
    print(f"Control: {control_success} retidos de {control_total} usu√°rios")
    
    # Executar teste Z
    stat, p_value = proportions_ztest(
        [target_success, control_success], 
        [target_total, control_total]
    )
    
    print(f"\nTeste Z: estat√≠stica = {stat:.4f}, p-value = {p_value:.6f}")
    print(f"Significativo (p < 0.05): {'SIM' if p_value < 0.05 else 'N√ÉO'}")
    
    # Interpreta√ß√£o
    print(f"\n--- INTERPRETA√á√ÉO ---")
    if p_value < 0.05:
        if lift_absoluto > 0:
            print(f"‚úÖ A campanha teve impacto POSITIVO e SIGNIFICATIVO")
            print(f"   Aumento de {lift_relativo:.2f}% na reten√ß√£o")
        else:
            print(f"‚ö†Ô∏è A campanha teve impacto NEGATIVO e SIGNIFICATIVO")
            print(f"   Redu√ß√£o de {abs(lift_relativo):.2f}% na reten√ß√£o")
    else:
        print(f"üîç A campanha N√ÉO teve impacto significativo na reten√ß√£o")
        print(f"   Varia√ß√£o de {lift_relativo:+.2f}% n√£o estatisticamente significativa")
    
    return {
        'target_retencao': target_retencao,
        'control_retencao': control_retencao,
        'lift_absoluto': lift_absoluto,
        'lift_relativo': lift_relativo,
        'p_value': p_value,
        'significativo': p_value < 0.05
    }

# Executar an√°lise
resultado_retencao = analisar_retencao(df)
#control_retencao = df[~df['is_target']]['retido_jan'].mean()
# An√°lise de m√©tricas secund√°rias
print("=== AN√ÅLISE DE OUTRAS M√âTRICAS ===")

# 1. Frequ√™ncia de Pedidos
target_freq = df[df['is_target']=='target']['num_pedidos_mes_1'].mean()
control_freq = df[df['is_target']=='control']['num_pedidos_mes_1'].mean()
print(f"Pedidos em Janeiro - Target: {target_freq:.2f}, Control: {control_freq:.2f}")

# 2. Valor Gasto (GMV)
target_gmv = df[df['is_target']=='target']['total_amount_mes_1'].mean()
control_gmv = df[df['is_target']=='control']['total_amount_mes_1'].mean()
print(f"GMV em Janeiro - Target: R$ {target_gmv:.2f}, Control: R$ {control_gmv:.2f}")

# 3. Varia√ß√£o na Frequ√™ncia
target_delta_freq = df[df['is_target']=='target']['delta_freq'].mean()
control_delta_freq = df[df['is_target']=='control']['delta_freq'].mean()
print(f"Œî Frequ√™ncia - Target: {target_delta_freq:.2f}, Control: {control_delta_freq:.2f}")
```
